```python
# Groupby is used to group the data with some statastics and categorical data
# data.groupby(" Categorical Column to which we want to group by")['Numerical Column'].Operation()
plot_data = pd.DataFrame(df.groupby('Education')['Income'].mean())
plot_data
```

```python
plot_data.reset_index(inplace=True) # Resets Index Values
```

```python
plot_data
```

## Bar Plot:
* Compare categorical data.
* Show distribution of data within different categories.
* Display rankings or comparisons between different groups.

```python
plot_data.plot(kind='bar',x='Education',y='Income') # Vertical Bar plot 
```

```python
# Seaborn Gives Much More Beautified Visual Representation of Data  
plt.figure(figsize=(10,8))
sns.barplot(data=plot_data,x='Income',y='Education',hue='Education',legend=True) # Horizontal Bar Plot - ( we only swapped columns between x and y )
```

## Pie Chart:
* .
Compare relative sizes of different categori
* Show proportions or percentages of a whole.
* Visualize composition or distribution of categorical data.es.

```python
plot_data.plot(kind='pie',y='Income',labels=plot_data['Education'])
```

## BOXPLOT
-  Display distribution and spread of continuous data.
- Identify outliers and variability within the data
-  Compare distributions of multiple groups.

```python
plt.boxplot(df['Income'])
```

```python
sns.boxplot(x= 'Education', y = 'Income', hue='Education',data = df)
plt.show()
```

## Histogram:
* Display distribution of continuous data.
* Identify the frequency or count of data within specific intervals or bins.
* Detect outliers or unusual patterns in data.

```python
plt.hist(df['Income'], bins = 15)
plt.show()
```

```python
plt.hist(df['Income'], bins = 15, edgecolor='black')
plt.show()
```

```python
sns.histplot(data=df['Income'],kde=True) # Kde shows Density plot line 
```

## Heatmap:
* Display matrix-like data using colors.
* Identify patterns or correlations in multivariate data.
* Visualize density or concentration of data points.

```python
corr_data = df.drop(columns=['Education','Marital_Status','ID']).corr() # Shows Correlation factor 
# More the value better the correlation
# if value is Positive then its positive correlation - i.e. as x increases y also increases 
# if value is Negative then its negative correlation - i.e. as x increases y decreases  
corr_data
```

```python
sns.heatmap(corr_data,annot=True)
```

```python
# Now in above plot we can see that meat products has high correlation wiht Income we can also display it using scatter plot 
```

## Lineplot

- used to visualize linear data generally time series data 
- it displays trends and patterns in data over time 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('flights') #loading data from seaborn datastore 

# From the previous code
df['yearMonth'] = pd.to_datetime("01-"+df['month'].astype(str)+"-"+df['year'].astype(str))
df.set_index('yearMonth',inplace=True) # inplace will make the change permanent to the DF
print(df.head())

# Now plot
plt.figure(figsize=(10,5))
sns.lineplot(data=df,x=df.index,y=df.passengers)
plt.show()
```
